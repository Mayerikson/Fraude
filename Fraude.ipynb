# 1. Entendimento do Negócio


A empresa MYK , uma fintech especializada em processamento de transações financeiras, solicitou um estudo para analisar os riscos associados a fraudes e melhorar a eficiência operacional. A empresa deseja saber o seguinte:


1- Quais são os principais indicadores de risco (ex.: Distancia_localizacao_frequente, Emulador, App_adulterado) que contribuem para a detecção de fraudes, e como podemos ajustar nossos modelos para reduzir falsos positivos sem comprometer a segurança?

2-Como as decisões de recusa de transações impactam a experiência do usuário legítimo, e quais medidas podem ser tomadas para minimizar o impacto negativo de falsos positivos?


3-Quais são os vetores de ataque mais comuns (ex.: uso de emuladores, adulteração de apps, localizações falsas), e como podemos reforçar nossas defesas contra essas ameaças?

4-Como podemos otimizar o volume e o fluxo de transações para garantir eficiência operacional e minimizar gargalos no sistema?

5-Qual é o impacto financeiro das decisões de aprovação e recusa de transações, e como podemos maximizar a economia com a prevenção de fraudes enquanto minimizamos perdas com falsos positivos?

## 1.1 O que são Fraudes?



Fraude é um crime que busca obter vantagem sobre a vítima.

Tipos de Fraudes:
- Auto fraude.
- Fraude de identidade
- Falsidade ideológica
- Falsificação de comprovantes
- Validação de documentos

- Biometria
- Motor de regras
- Validação de e-mail / número de celular
- Validação do Device utilizado
- Geolocalização
- Validação de aplicativos “de risco” no celular (VPN, apps falsos)


**Dificuldades**

- Convencer a empresa a adotar ferramentas de prevenção a fraudes
(lucro / custo)

- Fraude é um evento raro, portanto tem que ser bem assertivo nas
regras para não impactar o cliente bom
- Os fraudadores se reinventam e aplicam golpes novos todos os dias, é
difícil acompanhar.

## 3.1 Análise Bidimensional e respostas as  perguntas de negócio


**1)Random Forest (Floresta Aleatória):**


É um "time de especialistas" (várias árvores de decisão) que votam juntos para tomar decisões.

Cada "especialista" analisa partes diferentes dos dados e características diferentes.

A decisão final é baseada na maioria dos votos (para classificação) ou na média (para regressão).

Por que usamos aqui? Porque é robusto contra overfitting, lida bem com dados desbalanceados (poucas fraudes) e mostra claramente quais fatores são mais importantes para detectar fraudes.



**2)Matriz de Confusão:**

É uma "tabela da verdade" que mostra:

Verdadeiros Positivos (TP): Fraudes detectadas corretamente

Falsos Positivos (FP): Transações normais bloqueadas por engano

Falsos Negativos (FN): Fraudas que passaram despercebidas

Verdadeiros Negativos (TN): Transações normais aprovadas corretamente

Por que usamos aqui? Porque em fraude não podemos olhar só acurácia - um falso negativo (deixar passar fraude) é muito mais grave que um falso positivo (bloquear transação legítima).




```python
# Remover colunas irrelevantes
df = df.drop(['timestamp_transacao', 'data', 'minutos', 'faixa_minutos'], axis=1, errors='ignore')

# Processar data_hora para extrair características úteis
df['hora'] = df['data_hora'].dt.hour
df['dia_semana'] = df['data_hora'].dt.dayofweek

# Definir features e target
features = [
    'Distancia_localizacao_frequente',
    'idade_device_dias',
    'emulador',
    'tem_localizacao_falsa',
    'app_adulterado',
    'valor_transacao',
    'decisao',
    'hora',
    'dia_semana'
]
target = 'Fraude'

# Definir X e y
X = df[features]
y = df[target]

# Dividir os dados
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Escalonar características numéricas
scaler = StandardScaler()
colunas_numericas = ['Distancia_localizacao_frequente', 'idade_device_dias', 'valor_transacao']
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()
X_train_scaled[colunas_numericas] = scaler.fit_transform(X_train[colunas_numericas])
X_test_scaled[colunas_numericas] = scaler.transform(X_test[colunas_numericas])

# Random Forest otimizado e leve
print("Treinando modelo Random Forest...")
rf = RandomForestClassifier(
    n_estimators=50,
    max_depth=8,
    min_samples_split=10,
    min_samples_leaf=5,
    class_weight='balanced',
    random_state=42,
    n_jobs=2
)

# Treinar o modelo
rf.fit(X_train_scaled, y_train)

# Avaliar o modelo
print("Avaliando modelo...")
y_pred = rf.predict(X_test_scaled)
y_proba = rf.predict_proba(X_test_scaled)[:, 1]

# Calcular métricas básicas
report = classification_report(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = cm.ravel()  # Definição explícita das variáveis aqui

print("\n===== DESEMPENHO DO MODELO RANDOM FOREST =====")
print(report)
print(f"Matriz de Confusão:\n{cm}")
print(f"Verdadeiros Positivos (TP): {tp}")
print(f"Falsos Positivos (FP): {fp}")
print(f"Verdadeiros Negativos (TN): {tn}")
print(f"Falsos Negativos (FN): {fn}")
print(f"Acurácia: {(tp+tn)/(tp+tn+fp+fn):.2%}")
print(f"Taxa de falsos positivos: {fp/(fp+tn):.2%}")
print(f"Taxa de falsos negativos: {fn/(fn+tp):.2%}")

# Importância das features
importancia_rf = pd.DataFrame({
    'Feature': features,
    'Importância': rf.feature_importances_
}).sort_values(by='Importância', ascending=False)

print("\n===== IMPORTÂNCIA DAS FEATURES =====")
display(importancia_rf)

# Visualizar importância das features
plt.figure(figsize=(10, 6))
sns.barplot(x='Importância', y='Feature', data=importancia_rf)
plt.title('Importância das Features no Random Forest')
plt.tight_layout()
plt.show()

# Preparar respostas para as perguntas
print("\n===== RESPOSTAS ÀS PERGUNTAS DA MYK =====")

# 1. Principais indicadores de risco
print("\nPergunta 1: Quais são os principais indicadores de risco e como reduzir falsos positivos?")
print(f"Os principais indicadores de risco identificados pelo modelo são:")
print(f"1. {importancia_rf.iloc[0]['Feature']} (importância: {importancia_rf.iloc[0]['Importância']:.3f})")
print(f"2. {importancia_rf.iloc[1]['Feature']} (importância: {importancia_rf.iloc[1]['Importância']:.3f})")
print(f"3. {importancia_rf.iloc[2]['Feature']} (importância: {importancia_rf.iloc[2]['Importância']:.3f})")

print("\nPara reduzir falsos positivos sem comprometer a segurança:")
print("- Ajustar o limiar de probabilidade para classificação (atualmente 0.5)")
print("- Implementar regras específicas para clientes com bom histórico")
print("- Criar diferentes níveis de segurança baseados no valor da transação")
print("- Atualizar o modelo regularmente com dados mais recentes")

# 2. Impacto na experiência do usuário
print("\nPergunta 2: Como as decisões de recusa impactam a experiência do usuário legítimo?")
print(f"- Taxa atual de falsos positivos: {fp/(fp+tn):.2%}")
print("- Impactos negativos: frustração, perda de confiança, possível abandono do serviço")

print("\nMedidas para minimizar o impacto:")
print("- Implementar verificação secundária rápida (SMS, biometria)")
print("- Oferecer suporte imediato para casos de recusa")
print("- Comunicar claramente os motivos da recusa")
print("- Criar um processo simplificado de contestação")

# 3. Vetores de ataque
print("\nPergunta 3: Quais são os vetores de ataque mais comuns e como reforçar defesas?")
print("Vetores de ataque identificados:")
print(f"- Emuladores (importância: {importancia_rf[importancia_rf['Feature'] == 'emulador']['Importância'].values[0]:.3f})")
print(f"- Aplicativos adulterados (importância: {importancia_rf[importancia_rf['Feature'] == 'app_adulterado']['Importância'].values[0]:.3f})")
print(f"- Falsificação de localização (importância: {importancia_rf[importancia_rf['Feature'] == 'tem_localizacao_falsa']['Importância'].values[0]:.3f})")

print("\nReforços recomendados:")
print("- Detecção avançada de emuladores via fingerprinting de dispositivo")
print("- Verificação de integridade do aplicativo")
print("- Verificação cruzada entre GPS, IP e redes móveis")
print("- Análise comportamental do usuário")

# 4. Otimização do fluxo
print("\nPergunta 4: Como otimizar o volume e fluxo de transações?")
print("Estratégias recomendadas:")
print("- Processamento em lote para transações de baixo risco")
print("- Sistema de cache para decisões sobre clientes recorrentes")
print("- Verificação em camadas: básica para todos, avançada apenas para casos suspeitos")
print("- Infraestrutura escalável para períodos de pico")
print("- Monitoramento em tempo real para detectar gargalos")

# 5. Impacto financeiro
print("\nPergunta 5: Qual é o impacto financeiro das decisões e como maximizar a economia?")
valor_medio_transacao = df['valor_transacao'].mean()
print(f"- Valor médio das transações: R$ {valor_medio_transacao:.2f}")
print(f"- Falsos positivos detectados: {fp} transações")
print(f"- Falsos negativos detectados: {fn} transações")

print("\nEstrategias para maximizar economia:")
print("- Ajustar limiares baseado no valor da transação")
print("- Implementar regras de negócio específicas para transações de alto valor")
print("- Segmentar clientes por nível de risco")
print("- Revisar e atualizar o modelo periodicamente")

print("\n===== CONCLUSÃO =====")
print(f"O modelo Random Forest demonstra ser eficaz para detecção de fraudes com {tp/(tp+fn):.2%} de taxa de detecção.")
print(f"Os três principais indicadores são: {importancia_rf.iloc[0]['Feature']}, {importancia_rf.iloc[1]['Feature']} e {importancia_rf.iloc[2]['Feature']}.")
print("Para implementação, recomenda-se ajustar os limiares por segmento de cliente e valor de transação.")
```

**3)Acurácia**

*O que é Acurácia?*

É o "placar geral" do modelo, mostrando quantas vezes ele acertou no total.

*Fórmula:*

Acurácia = (Verdadeiros Positivos + Verdadeiros Negativos) / Total de Previsões

Nesse caso: (3066 + 112294) / 119956 ≈ 96.17%

*Problema da Acurácia em Fraudes:*

Imagine um jogo onde:

96% das vezes você ganha sem fazer nada (transações normais).

4% são fraudes (e você precisa pegar).

Se um modelo sempre chutar "não é fraude", ele acerta 96% (parece ótimo, mas na verdade é inútil, pois nunca detecta fraudes!).



**4)Avaliando a Matriz de Confusão:**

Revocação (Recall) = Fraudes detectadas / Total de fraudes reais

## 3.2 Calcular impacto financeiro


Vamos calcular o impacto financeiro das decisões erradas de fraude, mostrando quanto dinheiro está sendo perdido potencialmente com falsos negativos (fraudes não detectadas) e o custo dos falsos positivos (transações legítimas recusadas):

**Obs: Essa dúvida eu respondi de forma básica no modelo e aqui eu preferi detalhar melhor as informações.**

















```python
# Calcular impacto financeiro
valor_medio_transacao = df['valor_transacao'].mean()

# Falsos negativos: fraudes não detectadas (perda direta)
perda_falsos_negativos = fn * valor_medio_transacao

# Falsos positivos: transações legítimas recusadas (perda indireta + custo operacional)
# Assumindo que 70% dos clientes afetados desistem da transação
taxa_conversao_perdida = 0.7
perda_falsos_positivos = fp * valor_medio_transacao * taxa_conversao_perdida

# Custo total potencial
custo_total = perda_falsos_negativos + perda_falsos_positivos

print("\n===== IMPACTO FINANCEIRO DETALHADO =====")
print(f"1. Perda por fraudes não detectadas (falsos negativos):")
print(f"   - {fn} transações fraudulentas aprovadas")
print(f"   - Valor médio: R$ {valor_medio_transacao:.2f}")
print(f"   - Perda total: R$ {perda_falsos_negativos:,.2f}")

print(f"\n2. Perda por transações legítimas recusadas (falsos positivos):")
print(f"   - {fp} transações legítimas recusadas")
print(f"   - Valor médio: R$ {valor_medio_transacao:.2f}")
print(f"   - Assumindo {taxa_conversao_perdida:.0%} de desistência: R$ {perda_falsos_positivos:,.2f}")

print(f"\n3. Custo total potencial: R$ {custo_total:,.2f}")

print("\n===== ESTRATÉGIAS PARA MAXIMIZAR ECONOMIA =====")
print("1. Ajustar limiares de decisão baseado no valor da transação:")
print("   - Limiar mais rigoroso para transações de alto valor")
print("   - Limiar mais relaxado para transações de baixo valor")

print("\n2. Implementar análise de custo-benefício:")
print("   - Custo da fraude vs. custo da verificação adicional")
print("   - Balancear segurança e experiência do usuário")

print("\n3. Segmentação de clientes:")
print("   - Regras diferenciadas para clientes com bom histórico")
print("   - Verificação reforçada para novos clientes")

print("\n4. Monitoramento contínuo:")
print("   - Acompanhar a evolução dos padrões de fraude")
print("   - Recalibrar o modelo regularmente")
```

```python
def gerar_pdf_progresso_financeiro():
    # Configuração do PDF com margens maiores
    pdf_pages = matplotlib.backends.backend_pdf.PdfPages("Progresso_Mensal_Ajustado.pdf")
    fig = plt.figure(figsize=(14, 12))

    # Dados da tabela (exatamente como na imagem)
    dados = [
        ["1", "Custo Total", "R$ 649k", "Redução 15%", "R$ 552k (-15%)", "Limiares dinâmicos reduziram falsos negativos"],
        ["1", "Falsos Positivos", "2.65%", "<2%", "1.9%", "Melhora, mas ainda impacta receita"],
        ["2", "Recall", "66%", "75%", "72%", "Segmentação melhorou detecção"],
        ["2", "Economia (R$)", "-", "R$ 195k", "R$ 210k", "Whitelist reduziu falsos positivos"],
        ["3", "Custo Total", "R$ 552k", "R$ 454k (-30%)", "R$ 440k (-32%)", "Meta superada! Modelo + regras manuais"]
    ]

    # Configuração da tabela principal
    ax = plt.subplot2grid((3, 1), (0, 0), rowspan=2)
    ax.axis('off')

    # Criar tabela com ajustes de largura de coluna
    tabela = ax.table(
        cellText=dados,
        colLabels=["Mês", "Métrica", "Valor Inicial", "Meta", "Valor Alcançado", "Interpretação"],
        cellLoc='left',
        loc='center',
        colWidths=[0.05, 0.15, 0.12, 0.12, 0.15, 0.4]  # Ajustes precisos de largura
    )

    # Formatação avançada
    tabela.auto_set_font_size(False)
    tabela.set_fontsize(9)  # Fonte um pouco menor para caber tudo
    tabela.scale(1, 1.5)   # Ajuste de escala vertical

    # Estilização das células
    for (i, j), cell in tabela.get_celld().items():
        cell.set_edgecolor('#555555')
        cell.set_linewidth(0.7)
        if i == 0:  # Cabeçalho
            cell.set_facecolor('#2c3e50')
            cell.set_text_props(color='white', weight='bold')
        else:
            cell.set_text_props(fontsize=9)
            # Destacar linhas com metas superadas
            if "Meta superada" in dados[i-1][5]:
                cell.set_facecolor('#e8f8f5')

    # Adicionar espaço extra abaixo da tabela
    plt.tight_layout(pad=3.0, h_pad=2.0)

    # Adicionar interpretação financeira
    ax_text = plt.subplot2grid((3, 1), (2, 0))
    ax_text.axis('off')
    texto = (
        "Interpretação Financeira:\n\n"
        "Mês 1: R$ 97k economizados com ajuste de limiares | Custo operacional +R$ 10k\n"
        "Mês 2: Whitelist evitou R$ 58k em falsos positivos\n"
        "Mês 3: Economia total de R$ 209k/mês (ROI 10:1)"
    )
    ax_text.text(0.05, 0.5, texto, ha='left', va='center', fontsize=10,
                bbox=dict(facecolor='#f8f9fa', edgecolor='#ddd', boxstyle='round,pad=1'))

    pdf_pages.savefig(fig, bbox_inches='tight')
    pdf_pages.close()
    print("PDF 'Progresso_Mensal_Ajustado.pdf' gerado com sucesso!")

gerar_pdf_progresso_financeiro()
```
