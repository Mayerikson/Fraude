# 1. Entendimento do Negócio


A empresa MYK , uma fintech especializada em processamento de transações financeiras, solicitou um estudo para analisar os riscos associados a fraudes e melhorar a eficiência operacional. A empresa deseja saber o seguinte:


1- Quais são os principais indicadores de risco (ex.: Distancia_localizacao_frequente, Emulador, App_adulterado) que contribuem para a detecção de fraudes, e como podemos ajustar nossos modelos para reduzir falsos positivos sem comprometer a segurança?

2-Como as decisões de recusa de transações impactam a experiência do usuário legítimo, e quais medidas podem ser tomadas para minimizar o impacto negativo de falsos positivos?


3-Quais são os vetores de ataque mais comuns (ex.: uso de emuladores, adulteração de apps, localizações falsas), e como podemos reforçar nossas defesas contra essas ameaças?

4-Como podemos otimizar o volume e o fluxo de transações para garantir eficiência operacional e minimizar gargalos no sistema?

5-Qual é o impacto financeiro das decisões de aprovação e recusa de transações, e como podemos maximizar a economia com a prevenção de fraudes enquanto minimizamos perdas com falsos positivos?

## 1.1 O que são Fraudes?



Fraude é um crime que busca obter vantagem sobre a vítima.

Tipos de Fraudes:
- Auto fraude.
- Fraude de identidade
- Falsidade ideológica
- Falsificação de comprovantes
- Validação de documentos

- Biometria
- Motor de regras
- Validação de e-mail / número de celular
- Validação do Device utilizado
- Geolocalização
- Validação de aplicativos “de risco” no celular (VPN, apps falsos)


**Dificuldades**

- Convencer a empresa a adotar ferramentas de prevenção a fraudes
(lucro / custo)

- Fraude é um evento raro, portanto tem que ser bem assertivo nas
regras para não impactar o cliente bom
- Os fraudadores se reinventam e aplicam golpes novos todos os dias, é
difícil acompanhar.

Oportunidades (Opportunities)

Melhor segmentação de clientes para aplicação de regras de segurança apropriadas

Ameaças (Threats)

Evolução constante das técnicas de fraude
Crescente sofisticação de emuladores e ferramentas de adulteração de apps
Expectativas elevadas dos clientes quanto à experiência de uso
Pressão competitiva para reduzir fricção nas transações

# 3  Preparação dos Dados

### 3.1 Ajustando Parâmetros

 **1.Feature Engineering**

O feature engineering é o processo de criar ou transformar variáveis (features) para melhorar o desempenho do modelo. No seu código, você realizou as seguintes transformações:

*a) Features Temporais*

hora : Extraiu a hora da transação (df['data_hora'].dt.hour).
Isso ajuda a identificar padrões relacionados ao horário das transações, como fraudes mais comuns à noite.
dia_semana : Extraiu o dia da semana (df['data_hora'].dt.dayofweek).

Isso pode revelar padrões semanais, como fraudes mais frequentes em dias específicos.
fim_de_semana : Criou uma flag binária (1 para sábado ou domingo, 0 para outros dias). Transações em fins de semana podem ser mais suspeitas.
madrugada : Identificou transações feitas entre meia-noite e 5 horas da manhã. Este é um horário incomum para muitas atividades legítimas.

*b) Features de Interação*

valor_por_idade : Calculou a razão entre o valor da transação e a idade do dispositivo (df['valor_transacao'] / (df['idade_device_dias'] + 1)).

Isso ajuda a identificar transações atípicas para dispositivos novos ou antigos.
valor_x_distancia : Multiplicou o valor da transação pela distância da localização frequente (df['valor_transacao'] * (df['Distancia_localizacao_frequente'] + 1)). Combinações de alto valor e grande distância podem ser indicadores de fraude.

*c) Features de Dispositivo*

suspeita_device : Criou uma flag binária que indica se o dispositivo tem comportamentos suspeitos (emulador, tem_localizacao_falsa, app_adulterado). Isso concentra informações sobre riscos relacionados ao dispositivo.

*d) Categorização de Valor da Transação*

faixa_valor : Dividiu os valores das transações em faixas baseadas nos quartis (25%, 50%, 75%). Isso permite que o modelo capture padrões relacionados ao tamanho da transação sem depender diretamente de valores absolutos.

**2. Modelos Utilizados:**

Foi utilizado dois componentes principais neste projeto:

*a) SMOTETomek para Balanceamento de Classes*

O SMOTETomek é uma técnica combinada de oversampling (SMOTE) e undersampling (Tomek Links). Ele foi usado para lidar com o desequilíbrio entre classes (fraude vs. não-fraude).

Por que foi necessário?

Dados de fraude geralmente são altamente desbalanceados, com muito mais transações legítimas do que fraudulentas.
O balanceamento melhora o desempenho do modelo ao garantir que ele aprenda padrões tanto das classes minoritárias (fraudes) quanto das majoritárias (não-fraudes).

b) LightGBM (Gradient Boosting Decision Trees)

O LightGBM é um algoritmo de gradient boosting baseado em árvores de decisão. Ele é conhecido por sua eficiência computacional e capacidade de lidar com grandes volumes de dados.
Parâmetros principais ajustados:
objective='binary': Indica que o problema é de classificação binária (fraude ou não-fraude).
scale_pos_weight=3: Ajusta o peso da classe minoritária (fraude) para compensar o desequilíbrio.
n_estimators=300: Número de árvores no modelo.
learning_rate=0.03: Taxa de aprendizado, controlando o tamanho dos passos durante o treinamento.
max_depth=10: Profundidade máxima das árvores.
num_leaves=31: Número máximo de folhas em cada árvore.
Por que foi necessário?
O LightGBM é ideal para problemas de classificação desbalanceados, pois é rápido, escalável e capaz de capturar interações complexas entre features.
Ele também fornece métricas de importância de features, ajudando a interpretar o modelo.

**3. Como Interpretar os Resultados**

Os resultados fornecem várias métricas e insights importantes:
*
a) Métricas Gerais*

Threshold Ótimo (0.76) :
Este valor define o ponto de corte para classificar uma transação como fraudulenta. Um threshold mais alto reduz falsos positivos, mas pode aumentar falsos negativos.

Custo Total (R$ 314.280,66) :

Representa o custo combinado de falsos positivos (transações legítimas bloqueadas) e falsos negativos (fraudes não detectadas). Um custo baixo indica que o modelo está equilibrado.

AUC-ROC (0.8847) :

Mede a capacidade do modelo de distinguir entre classes. Um valor próximo de 1 indica excelente desempenho.

*b) Importância de Features*

As features mais importantes identificadas pelo modelo (valor_transacao, idade_device_dias, valor_por_idade, etc.) mostram quais variáveis têm maior impacto na detecção de fraudes. Isso ajuda a entender os principais indicadores de risco.


**4. Por Que Esses Componentes Foram Necessários?**

*a) Feature Engineering*

Sem feature engineering, o modelo teria dificuldade em capturar padrões complexos nos dados. Por exemplo:
Horários e dias da semana são variáveis temporais que, sozinhas, podem não ser suficientemente informativas.
A combinação de features (como valor_por_idade e valor_x_distancia) cria novas informações que ajudam o modelo a aprender relações mais sutis.

*b) SMOTETomek*

O desequilíbrio de classes é um problema comum em detecção de fraudes. Sem balanceamento, o modelo tenderia a favorecer a classe majoritária (não-fraude), resultando em baixa sensibilidade.

*c) LightGBM*

O LightGBM foi escolhido porque:

É eficiente para grandes conjuntos de dados.

Lida bem com features categóricas e numéricas.
Fornece interpretabilidade através da importância de features.

```python
# Suprimir avisos
warnings.filterwarnings("ignore")

# Configurações para visualização
try:
    plt.style.use('seaborn-v0_8')  # Estilo mais recente do seaborn
except:
    plt.style.use('ggplot')  # Fallback para um estilo similar
sns.set_palette("colorblind")
pd.set_option('display.max_columns', None)

def calcular_custo(y_true, y_pred, valor_medio_transacao=176.42, custo_falso_positivo=10):
    """Calcula o custo total baseado em falsos positivos e falsos negativos."""
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    custo_falsos_positivos = fp * custo_falso_positivo
    prejuizo_falsos_negativos = fn * valor_medio_transacao
    custo_total = custo_falsos_positivos + prejuizo_falsos_negativos
    return custo_falsos_positivos, prejuizo_falsos_negativos, custo_total

def encontrar_threshold_otimo(y_true, y_proba, valor_medio_transacao=176.42, custo_falso_positivo=10):
    """Encontra o threshold que minimiza o custo total."""
    thresholds = np.arange(0.01, 1.0, 0.01)
    custos = []
    for threshold in thresholds:
        y_pred = (y_proba >= threshold).astype(int)
        _, _, custo_total = calcular_custo(y_true, y_pred, valor_medio_transacao, custo_falso_positivo)
        custos.append(custo_total)
    threshold_otimo = thresholds[np.argmin(custos)]
    return threshold_otimo, min(custos)

def criar_features_adicionais(df):
    """Cria features adicionais para melhorar o modelo."""
    # Features temporais
    df['hora'] = df['data_hora'].dt.hour
    df['dia_semana'] = df['data_hora'].dt.dayofweek
    df['fim_de_semana'] = df['dia_semana'].isin([5, 6]).astype(int)
    df['madrugada'] = ((df['hora'] >= 0) & (df['hora'] < 5)).astype(int)
    # Features de interação
    df['valor_por_idade'] = df['valor_transacao'] / (df['idade_device_dias'] + 1)
    df['valor_x_distancia'] = df['valor_transacao'] * (df['Distancia_localizacao_frequente'] + 1)
    # Features de device
    df['suspeita_device'] = ((df['emulador'] == 1) |
                            (df['tem_localizacao_falsa'] == 1) |
                            (df['app_adulterado'] == 1)).astype(int)
    # Categorização de valor da transação
    q25, q50, q75 = df['valor_transacao'].quantile([0.25, 0.5, 0.75])
    conditions = [
        (df['valor_transacao'] <= q25),
        (df['valor_transacao'] > q25) & (df['valor_transacao'] <= q50),
        (df['valor_transacao'] > q50) & (df['valor_transacao'] <= q75),
        (df['valor_transacao'] > q75)
    ]
    values = [0, 1, 2, 3]
    df['faixa_valor'] = np.select(conditions, values)
    return df

def preprocessar_dados(df):
    """Realiza pré-processamento dos dados."""
    # Remover colunas irrelevantes
    df = df.drop(['timestamp_transacao', 'data', 'minutos', 'faixa_minutos'], axis=1, errors='ignore')
    # Converter para datetime se necessário
    if 'data_hora' in df.columns and not pd.api.types.is_datetime64_any_dtype(df['data_hora']):
        df['data_hora'] = pd.to_datetime(df['data_hora'])
    # Adicionar features temporais e outras
    df = criar_features_adicionais(df)
    # Tratar valores nulos
    df['Distancia_localizacao_frequente'] = df['Distancia_localizacao_frequente'].fillna(-1)
    df['emulador'] = df['emulador'].fillna(0)
    df['tem_localizacao_falsa'] = df['tem_localizacao_falsa'].fillna(0)
    df['app_adulterado'] = df['app_adulterado'].fillna(0)
    return df

def treinar_modelo_otimizado(X_train, y_train, X_test, y_test):
    """Treina um modelo LightGBM otimizado."""
    # Aplicar SMOTETomek para balanceamento
    smote_tomek = SMOTETomek(sampling_strategy=0.25, random_state=42)
    X_train_balanced, y_train_balanced = smote_tomek.fit_resample(X_train, y_train)
    print(f"Distribuição após SMOTETomek: {Counter(y_train_balanced)}")
    # Treinar LightGBM com parâmetros otimizados
    lgbm = LGBMClassifier(
        objective='binary',
        boosting_type='gbdt',
        scale_pos_weight=3,
        n_estimators=300,
        max_depth=10,
        num_leaves=31,
        learning_rate=0.03,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_alpha=0.1,
        reg_lambda=0.3,
        min_child_samples=20,
        random_state=42,
        n_jobs=-1
    )
    lgbm.fit(X_train_balanced, y_train_balanced)
    return lgbm

def avaliar_modelo(modelo, X_test, y_test, valor_medio_transacao=176.42):
    """Avalia o modelo e retorna métricas e visualizações."""
    # Prever probabilidades
    y_proba = modelo.predict_proba(X_test)[:, 1]
    # Encontrar threshold ótimo baseado em custo
    threshold_otimo, custo_minimo = encontrar_threshold_otimo(y_test, y_proba, valor_medio_transacao)
    print(f"Threshold ótimo (baseado em custo): {threshold_otimo:.2f}")
    # Avaliar com threshold ótimo
    y_pred_otimo = (y_proba >= threshold_otimo).astype(int)
    # Métricas
    cm = confusion_matrix(y_test, y_pred_otimo)
    tn, fp, fn, tp = cm.ravel()
    custo_fp, prejuizo_fn, custo_total = calcular_custo(y_test, y_pred_otimo, valor_medio_transacao)
    print("\n===== DESEMPENHO DO MODELO (THRESHOLD OTIMIZADO) =====")
    print(classification_report(y_test, y_pred_otimo))
    print(f"Matriz de Confusão:\n{cm}")
    print(f"Recall: {tp / (tp + fn):.2%}")
    print(f"Taxa de Falsos Positivos: {fp / (fp + tn):.2%}")
    print(f"AUC-ROC: {roc_auc_score(y_test, y_proba):.4f}")
    # Análise de custos
    print("\n===== ANÁLISE DE CUSTOS =====")
    print(f"Custo com falsos positivos: R$ {custo_fp:,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.'))
    print(f"Prejuízo com falsos negativos: R$ {prejuizo_fn:,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.'))
    print(f"Custo total: R$ {custo_total:,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.'))
    # Importância das features
    importancia = pd.DataFrame({
        'Feature': X_test.columns,
        'Importância': modelo.feature_importances_
    }).sort_values(by='Importância', ascending=False)
    print("\n===== IMPORTÂNCIA DAS FEATURES =====")
    print(importancia.head(10))
    # Gerar visualizações
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Importância', y='Feature', data=importancia.head(10))
    plt.title('Top 10 Features Mais Importantes')
    plt.show()
    return importancia, threshold_otimo

def gerar_recomendacoes(importancia):
    """Gera recomendações baseadas na análise do modelo."""
    print("\n===== RECOMENDAÇÕES PARA MELHORAR A DETECÇÃO DE FRAUDES =====")
    top_features = importancia.head(5)['Feature'].tolist()
    print("\n1. INDICADORES DE RISCO MAIS IMPORTANTES")
    for i, feature in enumerate(top_features, 1):
        print(f"   {i}. {feature}")
    print("\n2. ESTRATÉGIAS PARA REDUZIR FALSOS POSITIVOS")
    print("   - Implementar verificação em camadas baseada no nível de risco")
    print("   - Usar thresholds dinâmicos baseados em perfil do usuário")
    print("\n3. MITIGAÇÃO DE VETORES DE ATAQUE")
    print("   - Implementar detecção avançada de emuladores e apps adulterados")
    print("   - Monitorar padrões incomuns de localização e horários")

def main():
    # Verificar se o dataframe foi carregado
    if 'df' not in globals():
        print("\nErro: Nenhum dataframe 'df' encontrado.")
        print("Por favor, carregue seus dados conforme as instruções abaixo:")
        print("""
from google.colab import files
uploaded = files.upload()
import io
df = pd.read_csv(io.BytesIO(uploaded['seu_arquivo.csv']), parse_dates=['data_hora'])
""")
        return

    try:
        # Pré-processamento dos dados
        print("\nIniciando pré-processamento dos dados...")
        df_processado = preprocessar_dados(df)

        # Definir features e target
        features = [
            'Distancia_localizacao_frequente',
            'idade_device_dias',
            'emulador',
            'tem_localizacao_falsa',
            'app_adulterado',
            'valor_transacao',
            'decisao',
            'hora',
            'dia_semana',
            'fim_de_semana',
            'madrugada',
            'valor_por_idade',
            'valor_x_distancia',
            'suspeita_device',
            'faixa_valor'
        ]
        target = 'Fraude'

        # Verificar se todas as colunas existem
        missing_cols = [col for col in features + [target] if col not in df_processado.columns]
        if missing_cols:
            print(f"\nErro: As seguintes colunas não foram encontradas no dataframe: {missing_cols}")
            return

        # Separar dados
        X = df_processado[features]
        y = df_processado[target]

        # Dividir e escalar os dados
        print("\nDividindo e escalando os dados...")
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
        scaler = StandardScaler()
        colunas_numericas = ['Distancia_localizacao_frequente', 'idade_device_dias', 'valor_transacao',
                            'valor_por_idade', 'valor_x_distancia']
        X_train[colunas_numericas] = scaler.fit_transform(X_train[colunas_numericas])
        X_test[colunas_numericas] = scaler.transform(X_test[colunas_numericas])

        # Treinar modelo
        modelo = treinar_modelo_otimizado(X_train, y_train, X_test, y_test)

        # Avaliar modelo
        importancia, threshold_otimo = avaliar_modelo(modelo, X_test, y_test)

        # Gerar recomendações
        gerar_recomendacoes(importancia)

        # Salvar modelo
        try:
            joblib.dump(modelo, 'modelo_lgbm_otimizado.pkl')
            joblib.dump(scaler, 'scaler.pkl')
            print("\nModelo e scaler salvos com sucesso.")
            # Criar link para download no Colab
            from google.colab import files
            files.download('modelo_lgbm_otimizado.pkl')
            files.download('scaler.pkl')
        except Exception as e:
            print(f"\nAviso: Não foi possível salvar o modelo: {e}")

    except Exception as e:
        print(f"\nOcorreu um erro durante a execução: {str(e)}")

if __name__ == "__main__":
    main()
```
